```python
#Importing the Libraries I need
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import random


##THIS IS THE NEURAL NET TRAINING
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
input_layer = tf.keras.layers.Flatten()
first_hidden_layer = tf.keras.layers.Dense(128, activation=tf.nn.relu)
second_hidden_layer = tf.keras.layers.Dense(128, activation=tf.nn.relu)
output_layer = tf.keras.layers.Dense(10, activation=tf.nn.softmax)

model = tf.keras.models.Sequential()
model.add(input_layer)
model.add(first_hidden_layer)
model.add(second_hidden_layer)
#This network has 4 layers, one input, 2 hidden, and one output
model.add(output_layer)

model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3)

val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_acc)


##THIS IS SHOWING EXAMPLES OF HOW IT ACTUALLY WORKS
predictions = model.predict([x_test])
#This line is me choosing a random number photo to try to guess what it is
num = random.randint(0,9999)

plt.imshow(x_test[num], cmap = plt.cm.binary)
plt.show()
#This is a prediction of which number it will be
np.argmax(predictions[num])
```
